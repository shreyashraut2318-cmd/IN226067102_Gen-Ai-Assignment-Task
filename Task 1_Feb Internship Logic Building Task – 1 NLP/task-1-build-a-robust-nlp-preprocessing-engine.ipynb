1. What is the difference between "Love" and "love" in NLP?
--
In NLP, the difference between "Love" and "love" is entirely dependent on whether our preprocessing pipeline is case-sensitive or case-insensitive.

If we haven't converted our text to lowercase (a process called Case Folding), the computer treats them as two completely different entities.
The Computer’s Perspective (ASCII/Unicode)
At the lowest level, computers don't see letters; they see numbers.

L has an ASCII value of 76.
l has an ASCII value of 108.

Because the binary codes are different, a model will treat them as two distinct "tokens" with no relationship to each other.


2. What happens if stopwords are not removed?
--
Increased "Noise" and Lower Accuracy
In most NLP tasks, the goal is to find the context. Stopwords like "the," "is," and "at" carry almost zero context.
The Problem: A simple model might think two documents are similar just because they both use the word "the" 50 times, even if one is about Space and the other is about Cooking.
The Result: Your model gets distracted by the "noise" and fails to focus on the "signal" (the important words like "Galaxy" or "Recipe").

High Dimensionality (The "Storage" Problem)
In Data Science, more data isn't always better if that data is redundant.
The Math: If your vocabulary has 10,000 unique words, but 200 of those are stopwords that appear in every single sentence, your feature matrix becomes massive.
The Result: Your model requires more RAM and more disk space to store vectors that don't actually help it learn.

Slower Training Times
Every extra word in your dataset is a calculation the computer has to make.
The Problem: If 40% of your dataset consists of "is," "and," and "the," your model is spending 40% of its training time processing words that don't change the outcome.
The Result: Your training loops take much longer to finish, which is a major headache during internship deadlines or university projects.

When should we actually KEEP them?
We wll find that modern deep learning has changed the rules. We should not remove stopwords if we are doing:

Natural Language Generation (LLMs): If we remove "the" or "is," the AI will talk like a broken robot ("I go store" instead of "I am going to the store").
Sentiment Analysis (Sometimes): Words like "not" or "but" are often in stopword lists, but they completely change the meaning of a sentence ("I am not happy").
Part-of-Speech Tagging: We need those small words to understand the grammar and structure of a sentence.


3. Provide two real-world scenarios where removing stopwords can be harmful.
--
Sentiment Analysis (Negation & Context) :
In sentiment analysis, stopwords like "not," "no," "but," or "never" are the "switches" that flip the entire meaning of a sentence.

Original Sentence: "The food was not good, but the service was great."
After Stopword Removal: "food good service great"
The Harm: The model now thinks the user had a 100% positive experience. By removing the "not," you've completely inverted the sentiment of the first half of the sentence. The word "but" also acts as a crucial structural marker that signals a shift in the speaker's opinion.


Legal or Medical Document Analysis :
In fields like law or medicine, small "function words" define the boundaries of permissions, symptoms, or orders.

Scenario (Legal): "The defendant is liable for the damages." vs. "The defendant is not liable for the damages."
After Stopword Removal: Both sentences become "defendant liable damages".
The Harm: In a legal setting, this is a catastrophic failure. Stopwords like "is," "for," "to," and "from" define relationships between entities (who owes what to whom). Removing them turns a precise legal instruction into a vague collection of nouns.


4. Explain the difference between stemming and lemmatization.
--
Stemming (The Rough Cutter) :
Stemming works by following a fixed set of rules to chop off the ends of words (suffixes). It doesn't actually "understand" the language; it just looks for common patterns like -ing, -ed, or -es.

How it works: It uses algorithms like the Porter Stemmer.
Speed: Very fast (because it’s just a "search and chop" rule).
Result: It often produces "stems" that aren't real words (e.g., "universal" becomes "univers").


Lemmatization (The Professional Editor) :
Lemmatization is much smarter. It looks at a dictionary (a Lexicon) and considers the context of the word—specifically its Part of Speech (POS)—to find the "Lemma" (the dictionary base form).

How it works: It checks if a word is a verb, noun, or adjective before reducing it.
Speed: Slower (because it has to look things up in a dictionary).
Result: It always produces a real, meaningful word.

Task 2:Build Advanced Preprocessing Function 

```python
import re

def preprocess_text(text):
    # --- ERROR HANDLING: Check if input is empty or null ---
    if not text or not isinstance(text, str) or text.strip() == "":
        return [], ""


    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Lowercase
    text = text.lower()

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    tokens = text.split()

    # Remove short words except 'no', 'not'
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence



message=["I have 2 dogs","This is   good","soooo goooood!!!","WOW!!! This is GREAT!!!","Visit http://example.com now"  ]
tokens=[]
cleaned_sentence=[]
for text in message:
    token,sentence=preprocess_text(text)
    tokens.append(token)
    cleaned_sentence.append(sentence)
print(f"Tokens: {tokens}")
print(f"cleaned_sentence: {cleaned_sentence}")
```

Task 6: Build Full Pipeline 

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.append(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

text_list=["This is THE best course!!!", 
    "Visit https://openai.com now!", 
    "Nooooo this is baaad!!!", 
    "OK OK OK I got it", ]
full_pipeline(text_list)
```
